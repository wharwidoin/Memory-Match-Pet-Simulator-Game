````python
llm = Replicate(
    model="ibm-granite/granite-8b-code-instruct-128k",
    model_kwargs={
        "top_k": 5,
        "top_p": 1.0,
        "max_new_tokens": 2048,
        "temperature": 0.7
    }
)

prompt = """
Fix and extend this virtual pet game to allow full player control through text-based prompts (like a menu). Each day, the player should be able to choose actions such as feeding, playing, or cleaning the pet. The pet ages daily, and hunger increases while cleanliness decreases gradually. The game ends if the pet dies. Show clear status updates and prompts to the user each day.

Here is the current code structure. Modify it to support interactive control and add input validation:

```python
class Pet:
    def __init__(self, name):
        self.name = name
        self.hunger = 5
        self.happiness = 5
        self.cleanliness = 5
        self.age = 0
        self.feed_amount = 1
        self.play_amount = 1
        self.clean_amount = 1
        self.age_amount = 1

    def feed(self):
        if self.hunger > 0:
            self.hunger -= self.feed_amount
            self.hunger = max(self.hunger, 0)

    def play(self):
        if self.happiness < 10:
            self.happiness += self.play_amount
        if self.hunger < 10:
            self.hunger += self.play_amount

    def clean(self):
        if self.cleanliness < 10:
            self.cleanliness += self.clean_amount
            self.cleanliness = min(self.cleanliness, 10)

    def pass_day(self):
        self.age += self.age_amount
        self.hunger += 1
        if self.cleanliness > 0:
            self.cleanliness -= self.age_amount / 2

    def status(self):
        return f"{self.name} is hungry {self.hunger:.1f}, happy {self.happiness}, and clean {self.cleanliness:.1f}."

    def is_alive(self):
        return self.hunger < 10 and self.cleanliness > 0

def main():
    pet = Pet("Fluffy")
    for day in range(10):
        pet.feed()
        pet.play()
        pet.clean()
        pet.pass_day()
        if not pet.is_alive():
            print(f"Pet {pet.name} died on day {day + 1}.")
            break
        print(pet.status())

if __name__ == "__main__":
    main()


"""
response = llm.invoke(prompt)
print(response)
````
